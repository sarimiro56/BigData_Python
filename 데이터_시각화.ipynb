```python
import numpy as np

plt.plot(np.random.randn(10), 'k', label='중학생') #검은색 실선
plt.plot(np.random.randn(10)*3, 'r--', label='고등학생') #빨간색 파선
plt.plot(np.random.randn(10)*10, 'g.', label='대학생') #선 없이 초록색 점 마커로 표시
plt.legend()
plt.show()
```

```python
plt.subplot(2, 2, 1) # 2행 2열 중 첫 번째 그래프
plt.plot(np.random.randn(10), 'b--')
plt.subplot(2, 2, 2) # 2행 2열 중 두 번째 그래프
plt.plot(np.random.randn(100), 'r-+', alpha=0.2)
plt.subplot(2, 2, 3) # 2행 2열 중 세 번째 그래프
plt.plot(np.random.randn(10), 'y^')
plt.subplot(2, 2, 4) # 2행 2열 중 네 번째 그래프
plt.plot(np.random.randn(10), 'g.')
plt.show()
```

```python
!pip install konlpy

from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from konlpy.tag import Okt
import numpy as np
```

```python
with open('text12.txt', 'r', encoding='utf-8') as f:
  text = f.read()
okt = Okt()
nouns = okt.nouns(text)
words = [n for n in nouns if len(n) > 1]
c = Counter(words)
print(c)
```

```python
wc = WordCloud(font_path='/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf',\
               width=400, height=400, scale=2.0, max_font_size=250)
gen = wc.generate_from_frequencies(c)
plt.figure()
plt.imshow(gen)

wc.to_file('텍스트_워드클라우드.png')
```
