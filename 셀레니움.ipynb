```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
```

```python
service = Service(executable_path=r'/usr/bin/chromedriver')
options = webdriver.ChromeOptions()
options.add_argument('--headless') #창이 나타나지 않도록 Headless 설정하기
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
#options.add_argument('--blink-settings=imagesEnabled=false') #이미지 안뜨게
driver = webdriver.Chrome(service=service, options=options)
```

```python
url = 'http://ncov.kdca.go.kr/'
driver.get(url)
```

```python
topnews = driver.find_element('xpath', '//*[@id="content"]/div[2]/div/section/div[2]/ul')

print(topnews.text)
```

```python
news = []
for i in range (1, 100):
  try:
    topnews = driver.find_element('xpath', '//*[@id="content"]/div[2]/div/section/div[2]/ul/li[' + str(i) + ']/a/p[1]')
  except:
    break
  news.append(topnews.text)
print(news)
```
